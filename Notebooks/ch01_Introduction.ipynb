```python
import numpy as np

np.random.seed(52)

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_theme()
sns.set_context("paper")
sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
```

```python
from src.dataset import generate_toy_dataset

# sin function
def sin(x: np.ndarray) -> np.ndarray:
    return np.sin(2 * np.pi * x)

domain = (0, 1)
xtrain, ytrain = generate_toy_dataset(sin, domain, 10, 0.3)

xtest = np.linspace(0, 1, 100)
ytest = sin(xtest)

plt.scatter(xtrain, ytrain, facecolor="none", edgecolor="b", s=50, label="Training data")
plt.plot(xtest, ytest, color="g", label="$\sin(2\pi x)$")
plt.xlabel("$x$", fontsize=14); plt.ylabel("$y$")
plt.legend(); plt.show()
```
